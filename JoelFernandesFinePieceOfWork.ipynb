```python
import pandas as pd
import numpy as np
```

```python
#Read CSV for Danny Green 

green_df = pd.read_csv("danny green stats .csv")

# delete unwanted colums

green_df = green_df.drop (columns=[
    'FT', 'FTA', 'FT%', 'ORB','DRB', 'TRB', 'AST', 'STL', 
    'BLK', 'TOV', 'PF', 'PTS', 'GmSc', '+/-','MP', 'FG','FGA','FG%', 'Tm', 'Rk', 'G'])

# headings
green_df = green_df.rename(columns={"Unnamed: 5":"Home vs Away","Unnamed: 7":"Win/Loss by:"})
green_df['Home vs Away'].fillna("Home", inplace = True)
green_df['Home vs Away'] = green_df['Home vs Away'].replace(to_replace = "@", value = "Away")
# STEP 5 - Add Player 
green_df['Name'] = "Danny Green"

green_mean = green_df['3P'].mean()
green_3_percentage = green_df['3P'] / green_df['3PA'] * 100
print(f" Danny Green on average, makes {round(green_mean,2)} 3 pointers per game")
print(f" Danny Green's 3 point percentage is {green_3_percentage.mean()}")
```

```python
#Read CSV for Kawai Lenord
leonard_df = pd.read_csv("kawai Lenord stats.csv")

leonard_df = leonard_df.drop (columns=[
    'FT', 'FTA', 'FT%', 'ORB','DRB', 'TRB', 'AST', 'STL', 
    'BLK', 'TOV', 'PF', 'PTS', 'GmSc', '+/-','MP', 'FG','FGA','FG%', 'Tm', 'Rk', 'G'])

#  Add Player 
leonard_df['Name'] = "Kawai Leonard"
leonard_df = leonard_df.rename(columns={"Unnamed: 5":"Home vs Away","Unnamed: 7":"Win/Loss by:"})
leonard_df['Home vs Away'].fillna("Home", inplace = True)
leonard_df['Home vs Away'] = leonard_df['Home vs Away'].replace(to_replace = "@", value = "Away")
# headings
leonard_df = leonard_df.rename(columns={" ":"Home vs Away"," .1":"Win/Loss by:"})


leonard_mean = leonard_df['3P'].mean()
leonard_3_percentage = leonard_df['3P'] / leonard_df['3PA'] * 100
print(f" Kawhi Leonard on average, makes {round(leonard_mean,2)} 3 pointers per game")
print(f" Kawhi Leonard's 3 point percentage is {leonard_3_percentage.mean()}")
```
