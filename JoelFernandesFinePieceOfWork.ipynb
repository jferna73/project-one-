```python
import pandas as pd
import numpy as np
```

```python
#Read CSV for Danny Green 

green_df = pd.read_csv("danny green stats .csv")

# delete unwanted colums

green_df = green_df.drop (columns=[
    'FT', 'FTA', 'FT%', 'ORB','DRB', 'TRB', 'AST', 'STL', 
    'BLK', 'TOV', 'PF', 'PTS', 'GmSc', '+/-','MP', 'FG','FGA','FG%', 'Tm', 'Rk', 'G'])

# headings
green_df = green_df.rename(columns={" ":"Home vs Away"," .1":"Win/Loss by:"})

# STEP 5 - Add Player 
green_df['Name'] = "Danny Green"

green_df
```

```python
#Read CSV for Kawai Lenord
leonard_df = pd.read_csv("kawai Lenord stats.csv")

leonard_df = leonard_df.drop (columns=[
    'FT', 'FTA', 'FT%', 'ORB','DRB', 'TRB', 'AST', 'STL', 
    'BLK', 'TOV', 'PF', 'PTS', 'GmSc', '+/-','MP', 'FG','FGA','FG%', 'Tm', 'Rk', 'G'])

#  Add Player 
leonard_df['Name'] = "Kawai Leonard"

# headings
leonard_df = leonard_df.rename(columns={" ":"Home vs Away"," .1":"Win/Loss by:"})



leonard_df
```

```python
#Read CSV for Marc Gasol

gasol_df = pd.read_csv("marc Gasol stats.csv")



gasol_df = gasol_df.drop (columns=[
    'FT', 'FTA', 'FT%', 'ORB','DRB', 'TRB', 'AST', 'STL', 
    'BLK', 'TOV', 'PF', 'PTS', 'GmSc', '+/-','MP', 'FG','FGA','FG%', 'Tm', 'Rk'])


#Player name 

gasol_df['Name'] = "Marc Gasol"


gasol_df
```

```python
player_list = [green_df,leonard_df,gasol_df]
new_df = pd.concat(player_list)
new_df.groupby(['Date'])
```
